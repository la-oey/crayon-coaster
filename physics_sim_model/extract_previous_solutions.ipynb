```python
import numpy as np
import pandas as pd
import json
import glob
import ast
```

```python
# open json file listing game1 images that are rated in game2
ratedImgsFile = open('../analysis/game1/dfs/imagesRate.json')
ratedImgs = json.load(ratedImgsFile)
ratedImgs
```

```python
ratedImgs['boomerang_middle-left']
```

```python
# gather unique subjectIDs from rated images
ratedimgs = []

for levelwind in ratedImgs:
    for i in ratedImgs[levelwind]:
        ratedimgs.append(i)
        
ratedimgs[:5]
```

```python
# unique subjects from rated images
uniqueSubjs = (
    pd.DataFrame({'image' : imgs})
    .assign(designer = lambda x: x['image'].str.split("_").str[0])
    .designer.unique()
)

uniqueSubjs[:5]
```

```python
# tie images to trial information

trialdf = pd.read_csv("../analysis/game1/imagedf2.csv")

imagetrialdf = (
    trialdf
    .query('imageID.isin(@ratedimgs)')
)

# is there not a more efficient way to handle case_when in pandas??
imagetrialdf.loc[imagetrialdf['gravX'] == -0.25, 'wind'] = 'left'
imagetrialdf.loc[imagetrialdf['gravX'] == 0.25, 'wind'] = 'right'
imagetrialdf.loc[imagetrialdf['gravX'] == 0.00, 'wind'] = 'none'
imagetrialdf.head()
```

# structure rated images information
level-wind: {
    ratedImage: {
        drawnLines: [line1, line2],
        physObjs: [ [circle, rect1, rect2], [circle, rect1, rect2] ]
    }
}

```python
# create dictionary (key: levelID) of dictionaries (key: imageID) of dictionaries (key:drawnLines/PhysObj) :P

imageStrokeDict = {}

for index,row in imagetrialdf.iterrows():
        
    levelID = row['levelID'] + "-" + row['wind']
    if levelID not in imageStrokeDict:
        imageStrokeDict[levelID] = {row['imageID'] : {
            'drawnLines' : ast.literal_eval(row['drawnLines']),
            'drawnPhysObj' : ast.literal_eval(row['drawnPhysObj'])
        }}
    else:
        imageStrokeDict[levelID][row['imageID']] = {
            'drawnLines' : ast.literal_eval(row['drawnLines']),
            'drawnPhysObj' : ast.literal_eval(row['drawnPhysObj'])
        }
   

with open('drawnLines.json', 'w', encoding='utf-8') as jsonf:
    jsonf.write(json.dumps(imageStrokeDict, indent=4))
```
